```
for i in range (100):
  FizzBuzz = ''
  if (i + 1) % 3 == 0:
    FizzBuzz += 'Fizz'
  if (i + 1 ) % 5 == 0:
    FizzBuzz += 'Buzz'
  if FizzBuzz == '':
    print(i + 1)
  else:
    print(FizzBuzz)
```
